```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#Importing CSV and creating 
df = pd.read_csv("/Users/hervinsagnep/Downloads/data.csv")
df = df.set_index(df.Counter)
```

```python
#Cleaning Number of likes column
#Removing " VIEWS" from any strings containing it
df['Number of likes'] = df['Number of likes'].str.replace(" VIEWS", '').astype(int)
```

```python
#Calculates Average Number of Likes
average_likes = np.sum(df["Number of likes"]) / len(df)
```

```python
#Replaces "YES" or "NO" with Booleans
d = {"YES": True,"NO":False}
to_change = list(df)[4:]
df.replace(d,inplace=True)
```

```python
#Converting "Moneth-year" column to Datetime
from datetime import datetime
df["Month-year"] = df["Month-year"].apply(lambda i: datetime.strptime(i,'%m-%y'))
#Setting "Month-year" Datetime column as index to convert into TimesSeries
df = df.set_index(df["Month-year"])
```

```python
plt.figure(figsize =(20,10))
plt.plot(df["Number of likes"])
plt.show()
```
